## <b> Data Preprocessing <b>

Import Libraries and Dataset

```python
# import dataset
import numpy as np
import pandas as pd
```

```python
# load dataset
dataset = pd.read_csv(r"georgia_rec_copy.csv")
dataset
```

```python
dataset.shape
```

```python
# lists column names
column_names = dataset.columns.values.tolist()
column_names
```

Find Missing Values

```python
# replaces missing values with 0s
dataset = dataset.fillna(0)
```

```python
# sum of any missing values
dataset.isnull().sum()
```

Encoding Categorical Data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
dataset.dtypes
```

```python
le.classes_
```
