## Import Dataset and Libraries

## Questions

Conditional Selection

```python
# return a specific column
df.Gender
```

```python
df.Race
```

Which inmates took 3 years to return to prison?

```python
df[df.Recidivism_Within_3years == df.Recidivism_Within_3years.max()]
```

How many inmates did not return to prison?

```python
df[df.Recidivism_Within_3years == df.Recidivism_Within_3years.min()]
```

```python
df[df.Recidivism_Within_3years == df.Recidivism_Within_3years.max()].Race
```

Which group of people is the majority?

```python
df[df.Race == df.Race.max()].Race
```

Out of the majority race, who returned to prison in 3 years?

```python
df[df.Race == df.Race.max()].Recidivism_Within_3years.max
```

Which inmate(s) are Offense Type Drug, Other, Property, Violent/Non-Sex, Violent/Sex?

```python
# from prison offense column, group by drug
df.groupby('Prison_Offense').get_group('Drug')
```

```python
# from prison offense column, group by other
df.groupby('Prison_Offense').get_group('Other')
```

```python
# from prison offense column, group by property
df.groupby('Prison_Offense').get_group('Property')
```

```python
# from prison offense column, group by violent/non-sex
df.groupby('Prison_Offense').get_group('Violent/Non-Sex')
```

```python
# from prison offense column, group by violent/sex
df.groupby('Prison_Offense').get_group('Violent/Sex')
```
