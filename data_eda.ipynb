## Histograms

```python
# data for histogram 
OR = df[['Offense_Type', 'Race_Ethnicity']]
# looking at offense type and race to see which races commit which crimes
offense_type = ['Violent', 'Property', 'Other', 'Drug', 'Public Order']
OR = OR[OR['Offense_Type'].isin(offense_type)]
OR
```

```python
# histogram visual
OR['Offense_Type'].hist(by = OR['Race_Ethnicity'])
```

```python
OR['Offense_Type'].hist(by=OR['Race_Ethnicity'], layout = (1,12))
```

## Seaborn Data Visualization

```python
# import libraries
import seaborn as sns 
import matplotlib.pyplot as plt 
# another way of viewing the above histogram
g = sns.FacetGrid(OR, col = "Race_Ethnicity")
g.map(plt.hist, "Offense_Type");
```
